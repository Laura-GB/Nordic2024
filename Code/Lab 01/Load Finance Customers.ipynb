```python
df = spark.read.format("csv").option("header","true").load("Files/Finance/Customers.csv")
# df now is a Spark DataFrame containing CSV data from "Files/Finance/Customers.csv".
display(df)
```

```python
# Code generated by Data Wrangler for PySpark DataFrame

from pyspark.sql import functions as F
from pyspark.sql import types as T

def clean_data(df):
    # Rename column 'Cust Number' to 'CustNumber'
    df = df.withColumnRenamed('Cust Number', 'CustNumber')
    # Derive column 'City' from column: 'Location'
    # Transform based on the following examples:
    #    Location              Output
    # 1: "Newport, England" => "Newport"
    pd_udf_fn = F.pandas_udf(lambda v : v.str.split(",").str[0], T.StringType())
    df = df.withColumn("City", pd_udf_fn(F.col("Location")))
    # Derive column 'Region' from column: 'Location'
    # Transform based on the following examples:
    #    Location              Output
    # 1: "Newport, England" => "England"
    pd_udf_fn = F.pandas_udf(lambda v : v.str.split(",").str[-1].str.strip(), T.StringType())
    df = df.withColumn("Region", pd_udf_fn(F.col("Location")))
    # Drop column: 'Location'
    df = df.drop('Location')
    return df

df_clean = clean_data(df)
display(df_clean)
```

```python
df_clean.write.mode("overwrite").format("delta").option("overwriteSchema", "true").save("Tables/Finance_Customers")
```
